# Importing Models

```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
df=pd.read_csv('placement_records.csv')
df.head()
```

```python
df
```

# Pre- Processing

```python
df.isnull().sum()
```

```python
df=df.drop(['salary'], axis = 1)
```

```python
df.isnull().sum()
```

# Data Visualization

```python
sns.countplot(x= 'status', data=df)
```

```python
sns.countplot(x='status', hue='gender', data=df)
```

# Convert categorical variable into indicator variables

```python
workexp = pd.get_dummies(df['workex'])
```

```python
workexp
```

# Visualizing the correlation

```python
dataplot = sns.heatmap(df.corr(), annot=True)
plt.show()
```

# Concantenate data into original dataset

```python
df=pd.concat([df,workexp],axis=1)
```

```python
# df.drop(['sl_no','gender','ssc_b','hsc_s','hsc_b','degree_t','workex','specialisation'],axis=1,inplace=True)
df=df.drop(['sl_no','gender','ssc_b','hsc_s','hsc_b','degree_t','workex','specialisation'],axis=1)
```

```python
df.head()
```

# Train Test Split

```python
X=df.drop('status',axis=1)
y=df['status']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
```

# Appling Naive Bayes theorem

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_train,y_train)
```

```python
# compare actual and predicted
y_pred=model.predict(X_test)
df1=pd.DataFrame({'Actual Status':y_test,'Predicted Status':y_pred})
```

```python
df1
```

```python
print(classification_report(y_test, y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(accuracy_score(y_test,y_pred)*100)
```

# Model Intepretation

### Why Naive Bayes for this dataset?
 - We are using Navie Bayes for this dataset as the datset is categorical and the independent variables (conditionally idependent) are unrelated with each other.
 
### Classification report interpretation:
- From the classification report we can see that the precision and recall for class "Not placed" is 0.75 and 0.90 respectively meaning that the 75% of the predictions made by the model is correct and 90% of the relevant data points were correctly identified. Similary for class "Placed" precision is 95% and recall is 87%. Means the preprocessed dataset is well trained and processed such that it yeilds about 95% precise outcome for classifying "Placed" and 75% average outcome for "Not Placed". The model is found to be highly accurate.
 - And as for the accuracy of the model it is 0.88 i.e., 88% of outcomes were predicted correct.

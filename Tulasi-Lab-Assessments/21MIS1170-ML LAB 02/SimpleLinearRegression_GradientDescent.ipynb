# Data Visualization

```python
X = df.iloc[:, 0]
Y = df.iloc[:, 1]
plt.scatter(X, Y)
plt.show()
```

```python
# Building the model
m = 0
c = 0

L = 0.0001  # The learning Rate
epochs = 1000  # The number of iterations to perform gradient descent

n = float(len(X)) # Number of elements in X

# Performing Gradient Descent 
for i in range(epochs): 
    Y_pred = m*X + c  # The current predicted value of Y
    D_m = (-2/n) * sum(X * (Y - Y_pred))  # Derivative wrt m
    D_c = (-2/n) * sum(Y - Y_pred)  # Derivative wrt c
    m = m - L * D_m  # Update m
    c = c - L * D_c  # Update c
    print (m, c)
```

```python
Y_pred = m*X + c

plt.scatter(X, Y)
plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='red') # predicted
plt.show()
```

Why did we useGradient Descentn ? - Gradient Descenton is useto find the values of a function's parameters (coefficients) that minimize a cost function as far as possiblee
 - It iteratively adjusts model parameters towards the direction of steepest descent of the cost function, allowing for efficient convergence to the optimal solution.
 - With a dataset correlating years of experience to salary, gradient descent helps the model find the best-fitting line by minimizing the difference between predicted and actual salaries, optimizing the predictive power of the model.

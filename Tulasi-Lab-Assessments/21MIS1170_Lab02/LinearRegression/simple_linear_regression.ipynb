## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.metrics import r2_score
```

## Importing the dataset using Pandas

```python
dataset = pd.read_csv('Salary_Data.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

## Splitting the dataset into the 70% Training set and 30% Test set using Sklearn

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

## Training the Linear Regression model on the Training set with LinearRegression function from Sklearn

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train, y_train)
```

## Predicting the Test set results

```python
y_pred = reg.predict(X_test)
```

## Visualising the Training set results using scatter plot

```python
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, reg.predict(X_train), color = 'blue')
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

## Visualising the Test set results using scatter plot

```python
plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_train, reg.predict(X_train), color = 'blue')
plt.title('Salary vs Experience (Test set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
# R-squared method

r2 = r2_score(y_test, y_pred) 
print('r2 score: ', r2)
```

### Why did we use Simple Linear Regression?
##### Simple Linear Regression is used to model a relationship between two continuous variables. Linear Regression describes the correlation between an independent and a possible dependent varibale. Here our dataset has 1 dependent and independent features and hence it will be the best suit.

- Coefficient i.e, the Slope of the line is 9360.2613, which means the change of Y for a unit of increase in X.
- The intercept represents the value of the dependent variable when all independent variables are zero. In the case of a salary vs. experience linear regression model here, the intercept of 26777.39 would represent the estimated starting salary for individuals with zero years of experience..

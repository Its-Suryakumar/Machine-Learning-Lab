### Importing necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import scipy.cluster.hierarchy as shc
from scipy.spatial.distance import squareform, pdist
```

### Generating the dataset

```python
a = np.random.random_sample(size = 15)
b = np.random.random_sample(size = 15)
point = ['P1','P2','P3','P4','P5','P6','P7','P8','P9','P10','P11','P12','P13','P14','P15']
```

```python
df = pd.DataFrame({'Point':point, 'a':np.round(a,2), 'b':np.round(b,2)})
df = df.set_index('Point')
df
```

```python
df.describe()
```

### Visualizing the data

```python
plt.figure(figsize=(8,5))
plt.scatter(df['a'], df['b'], c='b', marker='o')
plt.xlabel('Column a')
plt.ylabel('column b')
plt.title('Scatter Plot of x and y')
for j in df.itertuples():
    plt.annotate(j.Index, (j.a, j.b), fontsize=15)
```

```python
dist = pd.DataFrame(squareform(pdist(df[['a', 'b']]), 'euclidean'), columns=df.index, index=df.index)
dist
```

### Single Link Clustering

```python
def single_linkage(dist_matrix):
    n = len(dist_matrix)
    while n > 1:
        min_val = float('inf')
        min_index = None
        for i in range(n):
            for j in range(i+1, n):
                if dist_matrix.iloc[i, j] < min_val and dist_matrix.index[i] != min_index:
                    min_val = dist_matrix.iloc[i, j]
                    min_index = (i, j)
        if min_val == float('inf'):
            break
        i, j = min_index
        cluster1, cluster2 = dist_matrix.index[i], dist_matrix.columns[j]
        print(f'Merging clusters {cluster1} and {cluster2} with distance {min_val}')
        new_cluster = f'({cluster1},{cluster2})'
        dist_matrix[new_cluster] = dist_matrix[[cluster1, cluster2]].min(axis=1)
        dist_matrix.loc[new_cluster] = dist_matrix[[cluster1, cluster2]].min(axis=0)
        dist_matrix = dist_matrix.drop([cluster1, cluster2], axis=0)
        dist_matrix = dist_matrix.drop([cluster1, cluster2], axis=1)
        n -= 1
        print(dist_matrix)
        print("")

print("Single-Link Clustering:")
single_linkage(dist.copy())
```

As indicated in the output, the points (P1,P3) is having the the least distance of all with 0.036055, so
initially we'll merge those two points together considering the least distance between them, to cluster 
together. Then the next least distance is between P14 and (P1,P3) with distance 0.063245 , so we'll merge those two clusters.
Similarly, calculating the least distance between each points and clusters and merging them based on the 
minimum distance between the neighbouring clusters is called Single Linkage Clustering. This loop
continues until all data points are clustered, revealing patterns in the data.

#### Dendrogram for Single Linkage

```python
plt.figure(figsize=(12,5))
plt.title("Dendrogram with Single Linkage")
dend = shc.dendrogram(shc.linkage(df[['a', 'b']], method='single'), labels=df.index)
```

### Complete Link Clustering

```python
def complete_linkage(dist_matrix):
    n = len(dist_matrix)
    while n > 1:
        min_val = float('inf')
        min_index = None
        for i in range(n):
            for j in range(i+1, n):
                if dist_matrix.iloc[i, j] < min_val and dist_matrix.index[i] != min_index:
                    min_val = dist_matrix.iloc[i, j]
                    min_index = (i, j)
        if min_val == float('inf'):
            break
        i, j = min_index
        cluster1, cluster2 = dist_matrix.index[i], dist_matrix.columns[j]
        print(f'Merging clusters {cluster1} and {cluster2} with distance {min_val}')
        new_cluster = f'({cluster1},{cluster2})'
        dist_matrix[new_cluster] = dist_matrix[[cluster1, cluster2]].max(axis=1)
        dist_matrix.loc[new_cluster] = dist_matrix.loc[[cluster1, cluster2]].max(axis=0)
        dist_matrix = dist_matrix.drop([cluster1, cluster2], axis=0)
        dist_matrix = dist_matrix.drop([cluster1, cluster2], axis=1)
        n -= 1
        print(dist_matrix)
        print("")
print("Complete-Link Clustering:")
complete_linkage(dist.copy())
```

As indicated in the output, the points (P1,P3) is having the the least distance of all with 0.036055, so
initially we'll merge those two points together considering the least distance between them, to cluster 
together. Then the next least distance of all the maximum distances is between P6 and P11 with distance 
0.072111 , so we'll merge those two clusters. Similarly, calculating the least distance between each 
points and clusters and merging them based on the maximum distance between the neighbouring clusters
is called Complete Linkage Clustering. This loop continues until all data points are clustered, revealing 
patterns in the data.

#### Dendrogram for Complete Linkage

```python
plt.figure(figsize=(12,5))
plt.title("Dendrogram with Complete Linkage")
dend = shc.dendrogram(shc.linkage(df[['a', 'b']], method='complete'), labels=df.index)
```
